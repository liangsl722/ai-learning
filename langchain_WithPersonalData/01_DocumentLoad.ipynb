PDF文档
- pypdf: 处理pdf

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("data/MachineLearning-Lecture01.pdf")

pages = loader.load()
```

```python
print("文档被加载后储存在变量中，结构为：", type(pages))
```

```python
print("文档总页数：", len(pages))
```

```python
page = pages[0]
print(f"{type(page)=}")
```

```python
### 文档包含两个属性：
### page_content: 该文档页面的内容
### meta_data: 文档页面的元数据(描述性数据)

print(f"{page.page_content[0:500]=}")
print("")
print(f"{page.metadata=}")
```

YouTube 音频
- pydub：处理音频
- yt_dlp：下载音频

```python
from langchain_community.document_loaders.generic import GenericLoader
from langchain_community.document_loaders.parsers.audio import OpenAIWhisperParserLocal
from langchain_community.document_loaders.blob_loaders.youtube_audio import YoutubeAudioLoader
from faster_whisper import WhisperModel

url = "https://www.youtube.com/watch?v=_PHdzsQaDgw"
save_dir="data"

loader = GenericLoader(
    # 将链接url中的音频下载下来，保存到本地
    blob_loader=YoutubeAudioLoader(urls=[url], save_dir=save_dir),
    # 解析器将音频转化为文本
    # blob_parser=OpenAIWhisperParser(),
    blob_parser=OpenAIWhisperParserLocal()
)

docs = loader.load()
```

```python
print("Type of pages: ", type(docs))
print("Length of pages: ", len(docs))

page = docs[0]
print("Type of page: ", type(page))
print("Page_content: ", page.page_content[:500])
print("Meta Data: ", page.metadata)
```

```python
### 分割
from langchain_text_splitters import RecursiveCharacterTextSplitter

r_splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=20,
)

r_splitter.split_text(page.page_content)
```

网页文档：
- 处理网页链接urls

```python
from langchain_community.document_loaders import WebBaseLoader

url = "https://raw.githubusercontent.com/datawhalechina/d2l-ai-solutions-manual/master/docs/README.md"
header = {'User-Agent': 'python-requests/2.27.1',
          'Accept-Encoding': 'gzip, deflate, br',
          'Accept': '*/*',
          'Connection': 'keep-alive'}
loader = WebBaseLoader(web_path=url,header_template=header)

pages = loader.load()
```

```python
print("Type of pages: ", type(pages))
print("Length of pages: ", len(pages))

page = pages[0]
print("Type of page: ", type(page))
print("Page_content: ", page.page_content[:500])
print("Meta Data: ", page.metadata)
```
