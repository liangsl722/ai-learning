支持bool类型，进行索引选择

```python
x = torch.tensor([[1, 2, 3, 4, 5],[7,8,9,10,11]])
print(x,x.shape)
mask = x >2     # 生成bool索引
print(mask)

# bool索引进行筛选
# 本质是抽取所有满足条件的元素，并按一维返回
filtered_x = x[mask]
print(filtered_x, filtered_x.shape)

# 筛选并赋值
x[mask] = 0
print(x,x.shape)
```

tensor 形状变换

```python
# reshape() 元素个数需要一致，按元素顺序重新组织维度
x = torch.randn(4,4)
print(x, x.shape)
x = x.reshape(2,8)
print(x, x.shape)
```

```python
# permute() 交换维度
# 原序列 x.shape=(2,3,4) 执行permute(1, 2, 0)后, (2,3,4)→(3,4,2)
# 新序列的每个元素 y[a,b,c] = x[c, a, b]
x = torch.tensor([[1,2,3], [4,5,6]])
x_reshape = x.reshape(3, 2)
x_transpose = x.permute(1, 0)   # 交换第0个和第1个维度。对于二维矩阵就是行列交换，进行装置
print("reshape:",x_reshape)
print("permute:",x_transpose)
```

```python
# unsqueeze() 拓展维度，在指定位置插入一个长度为1的新维度
x = torch.tensor([[1,2,3], [4,5,6]])
print(x, x.shape)

# 扩展第0维
x_0 = x.unsqueeze(0)
print(x_0, x_0.shape)

# 扩展第1维
x_1 = x.unsqueeze(1)
print(x_1, x_1.shape)

# 扩展第2维
x_2 = x.unsqueeze(2)
print(x_2, x_2.shape)
```

```python
# squeeze() 缩减大小为1的维度
x = torch.ones((1,1,3))
print(x, x.shape)

# 删除所有size=1的维度
y = x.squeeze()
print(y, y.shape)

# 删除指定维度，如果其size=1
z = x.squeeze(dim=0)
print(z, z.shape)
```

统计函数

```python
x = torch.tensor(
    [
        [1,3],
        [1,3],
        [1,3],
    ],
    dtype=torch.float,
)

print(x, x.shape, x.dtype)

# mean()只支持浮点数或复数类型,进行连续计算，而不能对long或者int离散型进行连续统计量计算
# tensor指定统计的维度，意味着要消灭这个维度
# 指定dim=0，意味着要消灭行维度，最后得到的统计量只有列维度
print(f"{x.mean()=}")
print(f"{x.mean(dim=0)=}")
print(f"{x.mean(dim=1)=}")

print()
# 指定参数keepdim=True，保留原来的维度不变，不会消灭统计的维度
mean = x.mean(dim=0)
print("dim=0, keepdim=false: ", mean, mean.shape)
mean = x.mean(dim=0, keepdim=True)
print("dim=0, keepdim=false: ", mean, mean.shape)
```

索引和切片

```python
x = torch.tensor([[1,2,3], [4,5,6]])
print("原始：", x, x.shape)

x1 = x[0, 1]
print("第一行第二个元素：", x1, x1.shape)

x2 = x[:, 1]
print("第二列：", x2, x2.shape)     # 切片中，:相当于保留维度，整数索引相当于该维度被消灭掉

x3 = x[1, :]
print("第二行：", x3, x3.shape)

x4 = x[:2, :]
print("前两行：", x4, x4.shape)

x5 = x[:, :2]
print("前两列：", x5, x5.shape)
```

broadcasting 广播机制

```python
# 让shape不同的张量从最后一维开始对齐，然后进行逐元素运算
# 不是复制数据，而是通过调整张量的索引方式（类似于reshape）
# 原则：
# 维度对齐：从最后一维度开始对齐，在短的tensor前面补1
# 扩展维度：在维度值为 1 的维度上，虚拟复制，让两个tensor纬度值一样
# 按位计算：两个tensor要么维度值一样，要么不同的维度值中有一个是1


x = torch.ones((1,2,3))
y = torch.ones((2,2,1))

print(x+y, (x+y).shape)
```

利用GPU加速计算

```python
# 1.创建时，指定tensor的设备为 device="cuda"
# 2.将cpu上的tensor通过 to("cuda")方法转移到GPU上
# 3.通过 torch.cuda.is_available() 检查环境中是否有可能的cuda

import time

# 确保gpu可用
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"device: {device}")

# 生成随机矩阵
size = 10000
A_cpu = torch.rand(size, size)  # 默认在cpu上创建tensor
B_cpu = torch.rand(size, size)

# CPU上计算
start_cpu = time.time()
C_cpu = torch.mm(A_cpu, B_cpu)  # 矩阵乘法
end_cpu = time.time()
print(f"CPU time: {end_cpu-start_cpu:.6f} sec")

# GPU上计算
A_gpu = A_cpu.to(device)    # 将tensor转移到GPU上
B_gpu = B_cpu.to(device)

start_gpu = time.time()
C_gpu = torch.mm(A_gpu, B_gpu)
torch.cuda.synchronize()    # 等待GPU所有的核都完成计算
end_gpu = time.time()
print(f"GPU time: {end_gpu-start_gpu:.6f} sec")
```

自动求梯度

```python
x = torch.tensor(1.0, requires_grad=True)   # 指定需要计算梯度
y = torch.tensor(1.0, requires_grad=True)

# v = 3*x + 4*y
# u = torch.square(v)
# z = torch.log(u)
z = ((3*x+4*y)**2).log()

z.backward()    # 反向传播求梯度：对标量 z 进行反向传播，计算所有参与计算的张量的梯度

print("x grad: ", x.grad)
print("y grad: ", y.grad)
```

线性回归

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 生成数据
inputs = torch.rand(100, 3) # 在均匀分布上生成shape为(100,3)的tensor
weights = torch.tensor([[1.1], [2.2], [3.3]])   # 预设的权重
bias = torch.tensor(4.4)    # 预设的bias
targets = inputs @ weights + bias + 0.1*torch.randn(100,1)  # 增加随机误差

# 随机初始化线性回归的参数
# 初始化参数直接放在cuda上，并启用梯度追踪
w = torch.rand((3, 1), device=device, requires_grad=True)
b = torch.rand((1,), device=device, requires_grad=True)

# 训练
# 将数据移至相同设备
inputs = inputs.to(device)
targets = targets.to(device)

epoch = 10000
lr = 3e-3

for i in range(epoch):
    outputs = inputs @ w + b
    loss = torch.mean(torch.square(outputs - targets))

    if i % 1000 == 0:
        print(f"epoch {i}, loss: {loss.item()}")

    loss.backward()

    with torch.no_grad():   # 参数更新部分，不需要追踪梯度
        w -= lr * w.grad
        b -= lr * b.grad

    # 清零梯度
    w.grad.zero_()
    b.grad.zero_()

print("训练后的权重w：\n", w)
print("训练后的偏置b：\n", b)
```

loss 曲线: tensorboard 可视化训练过程

```python
from torch.utils.tensorboard import SummaryWriter

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 生成数据
inputs = torch.rand(100, 3)
weights = torch.tensor([[1.1], [2.2], [3.3]])
bias = torch.tensor(4.4)
targets = inputs @ weights + bias + 0.1 * torch.randn(100, 1)

# ****************** 创建一个SummaryWriter实例 ******************
writer = SummaryWriter(log_dir="//torch/runs/exp1")

# 初始化参数时直接放在CUDA上，并启用梯度追踪
w = torch.rand(3, 1, requires_grad=True, device=device)
b = torch.rand(1, requires_grad=True, device=device)

# 将数据移至相同设备
inputs = inputs.to(device)
targets = targets.to(device)

epoch = 10000
lr = 0.003

for i in range(epoch):
    outputs = inputs @ w + b
    loss = torch.mean(torch.square(outputs - targets))
    if i % 1000 == 0:
        print(f"epoch {i}, loss: {loss.item()}")

    # ******************记录loss，三个参数分别：tag，loss值，第几步******************
    writer.add_scalar("loss/train", loss.item(), i)

    loss.backward()

    with torch.no_grad():  # 下边的计算不需要跟踪梯度
        w -= lr * w.grad
        b -= lr * b.grad

    # 清零梯度
    w.grad.zero_()
    b.grad.zero_()

writer.close()
print("训练后的权重 w:", w)
print("训练后的偏置 b:", b)
```
